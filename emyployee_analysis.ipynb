```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file (adjust path if needed)
df = pd.read_csv('HR Data.xlsx - HR data.csv')
```

```python
# Remove spaces and standardize column names
df.columns = df.columns.str.strip().str.replace(' ', '')
```

```python
print(df.head())
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
print(df.describe(include='all'))
```

```python
df['Attrition'].value_counts().plot(kind='bar', color=['blue', 'red'])
plt.title('Employee Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
# Check the actual column names
print(df.columns)

# Plot only if column exists
if 'JobSatisfaction' in df.columns:
    df['JobSatisfaction'].plot(kind='hist', bins=4, edgecolor='black')
    plt.title('Job Satisfaction Distribution')
    plt.xlabel('Satisfaction Level')
    plt.ylabel('Frequency')
    plt.show()
else:
    print("Column 'JobSatisfaction' not found in the dataset.")
```

```python
# Age distribution histogram
df['Age'].plot(kind='hist', bins=10, edgecolor='black', color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
if 'Department' in df.columns:
    df['Department'].value_counts().plot(kind='bar', color='purple')
    plt.title('Employee Count by Department')
    plt.xlabel('Department')
    plt.ylabel('Count')
    plt.show()
else:
    print("Column 'Department' not found in the dataset.")
```

```python
if 'MonthlyIncome' in df.columns and 'Attrition' in df.columns:
    df.boxplot(column='MonthlyIncome', by='Attrition', grid=False)
    plt.title('Monthly Income by Attrition Status')
    plt.suptitle('')  # Remove the default title
    plt.xlabel('Attrition')
    plt.ylabel('Monthly Income')
    plt.show()
else:
    print("Column 'MonthlyIncome' or 'Attrition' not found in the dataset.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(25, 20))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap (Numeric Columns)')
plt.show()
```
